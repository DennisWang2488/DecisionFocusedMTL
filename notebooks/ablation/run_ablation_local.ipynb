# Ablation Experiments — Local Runner

This notebook executes `run_ablation.py` and displays the results.

**Experiments:**
1. **No continuation** — all methods reset weights per lambda stage
2. **Single lambda=0.5** — no cross-stage effects
3. **Fairness comparison** — MAD vs Gap vs Atkinson
4. **Predictor architecture** — linear, MLP, ResNet, FT-Transformer
5. **LR × steps grid** — convergence sensitivity
6. **Optimizer** — SGD vs Adam

**How to use:**
1. Run Cell 1 to execute experiments 1-3 (takes a few minutes)
2. Run remaining cells to load and display results
3. For HP experiments 4-6: `python run_ablation.py --experiments 4 5 6`

```python
# Cell 1: Run the ablation experiments
# Change n_sample, alpha, or experiments as needed

import subprocess, os, sys

PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), '..', '..'))
SCRIPT = os.path.join(PROJECT_ROOT, 'run_ablation.py')
RESULTS_DIR = os.path.join(PROJECT_ROOT, 'results', 'ablation')

N_SAMPLE = 1000
ALPHA = 2.0

cmd = [
    sys.executable, SCRIPT,
    '--n-sample', str(N_SAMPLE),
    '--alpha', str(ALPHA),
    '--results-dir', RESULTS_DIR,
    '--experiments', '1', '2', '3'
]

print(f'Running: {" ".join(cmd)}')
print(f'Results will be saved to: {RESULTS_DIR}')
print('=' * 70)

proc = subprocess.run(cmd, cwd=PROJECT_ROOT, capture_output=False)
if proc.returncode != 0:
    print(f'\nScript exited with code {proc.returncode}')
else:
    print('\nScript completed successfully')
```

```python
# Cell 2: Load results and setup

import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), '..', '..'))
RESULTS_DIR = os.path.join(PROJECT_ROOT, 'results', 'ablation')
ALPHA = 2.0

for p in [PROJECT_ROOT, os.path.join(PROJECT_ROOT, 'src')]:
    if p not in sys.path:
        sys.path.insert(0, p)

from configs import DEFAULT_TRAIN_CFG
from plotting import plot_gradient_conflict, plot_training_curves


def make_summary_table(sdf, group_cols=None):
    """Aggregate test metrics as mean +/- std across seeds.

    Reports **normalized regret** as the primary 'Regret' column (matching
    the main experiment notebooks).  Raw (unnormalized) regret remains in
    the underlying CSV logs.
    """
    if group_cols is None:
        group_cols = ['config_name']

    has_norm = 'test_regret_normalized' in sdf.columns
    regret_col = 'test_regret_normalized' if has_norm else 'test_regret'

    metrics = [regret_col, 'test_fairness', 'test_pred_mse']
    short = {
        'test_regret_normalized': 'Regret',
        'test_regret': 'Regret',
        'test_fairness': 'Fairness',
        'test_pred_mse': 'Pred MSE',
    }
    rows = []
    for keys, grp in sdf.groupby(group_cols):
        if isinstance(keys, str):
            keys = (keys,)
        row = dict(zip(group_cols, keys))
        for m in metrics:
            if m not in grp.columns:
                continue
            mu, sd = grp[m].mean(), grp[m].std()
            row[f'{short[m]}'] = f'{mu:.4f} +/- {sd:.4f}'
            row[f'{short[m]}_mean'] = mu
        row['n_seeds'] = len(grp)
        rows.append(row)
    return pd.DataFrame(rows)


# Load all CSV files
def load_csv(name):
    path = os.path.join(RESULTS_DIR, name)
    if os.path.exists(path):
        df = pd.read_csv(path)
        print(f'Loaded {name}: {len(df)} rows')
        return df
    print(f'NOT FOUND: {name}')
    return pd.DataFrame()

stage_df_nocont = load_csv('stage_no_continuation.csv')
iter_df_nocont = load_csv('iter_no_continuation.csv')
stage_df_single = load_csv('stage_single_lambda.csv')
iter_df_single = load_csv('iter_single_lambda.csv')

fairness_types = ['mad', 'gap', 'atkinson']
results_by_fairness = {}
for ft in fairness_types:
    s = load_csv(f'stage_fairness_{ft}.csv')
    i = load_csv(f'iter_fairness_{ft}.csv')
    if not s.empty:
        results_by_fairness[ft] = (s, i)

print(f'\nAll results loaded from {RESULTS_DIR}')
```

## Experiment 1: No Continuation

All methods have `continuation=False` — weights reset at each lambda stage.
Gradient cosines should start at the same point for all methods.

```python
# Cell 3: Exp 1 — gradient cosine and training curve plots

if not iter_df_nocont.empty:
    nocont_dir = os.path.join(RESULTS_DIR, 'no_continuation')
    os.makedirs(nocont_dir, exist_ok=True)
    plot_gradient_conflict(iter_df_nocont, alpha_values=[ALPHA], results_dir=nocont_dir)
    plot_training_curves(iter_df_nocont, alpha_values=[ALPHA], results_dir=nocont_dir)
else:
    print('No Experiment 1 data found.')
```

```python
# Cell 4: Exp 1 — summary table per lambda

if not stage_df_nocont.empty:
    print(f'=== Experiment 1: No Continuation (alpha={ALPHA}) ===')
    print(f'Each cell shows mean +/- std across seeds')
    print(f'Regret = normalized regret (raw regret available in CSV logs)\n')

    for lam in sorted(stage_df_nocont['lambda'].unique()):
        sub = stage_df_nocont[stage_df_nocont['lambda'] == lam]
        tbl = make_summary_table(sub)
        tbl = tbl.sort_values('Regret_mean')
        print(f'--- lambda = {lam} ---')
        display_cols = ['config_name', 'Regret', 'Fairness', 'Pred MSE']
        print(tbl[[c for c in display_cols if c in tbl.columns]].to_string(index=False))
        print()

    # Best lambda per method
    has_norm = 'test_regret_normalized' in stage_df_nocont.columns
    regret_col = 'test_regret_normalized' if has_norm else 'test_regret'

    print('--- Best lambda per method (min normalized regret + fairness) ---')
    best_rows = []
    for method in sorted(stage_df_nocont['config_name'].unique()):
        msub = stage_df_nocont[stage_df_nocont['config_name'] == method]
        per_lam = msub.groupby('lambda').agg(
            regret=(regret_col, 'mean'),
            fairness=('test_fairness', 'mean'),
            mse=('test_pred_mse', 'mean'),
        ).reset_index()
        if len(per_lam) == 0:
            continue
        r_max = max(per_lam['regret'].max(), 1e-12)
        f_max = max(per_lam['fairness'].max(), 1e-12)
        per_lam['score'] = per_lam['regret'] / r_max + per_lam['fairness'] / f_max
        best = per_lam.loc[per_lam['score'].idxmin()]
        best_rows.append({
            'method': method,
            'best_lambda': best['lambda'],
            'regret': f'{best["regret"]:.4f}',
            'fairness': f'{best["fairness"]:.4f}',
            'mse': f'{best["mse"]:.4f}',
        })
    print(pd.DataFrame(best_rows).to_string(index=False))
else:
    print('No Experiment 1 data found.')
```

## Experiment 3: Fairness Measure Comparison

Same methods under MAD, Gap, and Atkinson fairness metrics.
Tests whether method rankings are consistent across fairness measures.

```python
# Cell 7: Exp 3 — per-fairness-type tables + rank consistency

if results_by_fairness:
    print(f'=== Experiment 3: Fairness Measure Comparison (alpha={ALPHA}, lambda=0.5) ===')
    print(f'Each cell shows mean +/- std across seeds')
    print(f'Regret = normalized regret (raw regret available in CSV logs)\n')

    for ft in fairness_types:
        if ft not in results_by_fairness:
            continue
        sdf, _ = results_by_fairness[ft]
        tbl = make_summary_table(sdf)
        tbl = tbl.sort_values('Regret_mean')
        tbl.insert(0, 'Rank', range(1, len(tbl) + 1))

        print(f'{"="*80}')
        print(f'  Fairness type: {ft.upper()}')
        print(f'{"="*80}')
        display_cols = ['Rank', 'config_name', 'Regret', 'Fairness', 'Pred MSE']
        print(tbl[[c for c in display_cols if c in tbl.columns]].to_string(index=False))
        print()

    # Cross-fairness rank comparison
    print(f'{"="*80}')
    print('  Ranking Consistency Across Fairness Types (by normalized regret, 1 = best)')
    print(f'{"="*80}')

    rank_dict = {}
    for ft in fairness_types:
        if ft not in results_by_fairness:
            continue
        sdf, _ = results_by_fairness[ft]
        tbl = make_summary_table(sdf).sort_values('Regret_mean').reset_index(drop=True)
        for rank, (_, r) in enumerate(tbl.iterrows(), 1):
            method = r['config_name']
            if method not in rank_dict:
                rank_dict[method] = {}
            rank_dict[method][ft.upper()] = rank

    rank_table = pd.DataFrame.from_dict(rank_dict, orient='index')
    rank_table.index.name = 'method'
    rank_table['Avg Rank'] = rank_table.mean(axis=1)
    rank_table = rank_table.sort_values('Avg Rank')
    print(rank_table.to_string(float_format='%.1f'))

    print('\n--- Regret ordering (best -> worst) ---')
    for ft in fairness_types:
        if ft not in results_by_fairness:
            continue
        sdf, _ = results_by_fairness[ft]
        tbl = make_summary_table(sdf).sort_values('Regret_mean')
        ranking = list(tbl['config_name'].values)
        print(f'  {ft.upper():10s}: {" > ".join(ranking)}')
else:
    print('No Experiment 3 data found.')
```

```python
# Cell 8: Exp 3 — gradient conflict plots per fairness type

for ft in fairness_types:
    if ft not in results_by_fairness:
        continue
    _, idf = results_by_fairness[ft]
    ft_dir = os.path.join(RESULTS_DIR, f'fairness_{ft}')
    os.makedirs(ft_dir, exist_ok=True)
    print(f'\n--- Gradient conflict: {ft.upper()} ---')
    plot_gradient_conflict(idf, alpha_values=[ALPHA], results_dir=ft_dir)
    plot_training_curves(idf, alpha_values=[ALPHA], results_dir=ft_dir)
```

## Gradient Cosine Tables

Per-iteration gradient cosine and norm values from Experiment 2 (single lambda=0.5).

```python
# Cell 9: Gradient cosine table — per method, per iteration

cos_cols = ['cos_dec_pred', 'cos_dec_fair', 'cos_pred_fair']
norm_cols = ['grad_norm_dec', 'grad_norm_pred', 'grad_norm_fair', 'grad_norm_combined']
display_cols = ['iter'] + cos_cols + norm_cols

if not iter_df_single.empty:
    df = iter_df_single.copy()
    for method in sorted(df['config_name'].unique()):
        msub = df[df['config_name'] == method]
        available = [c for c in display_cols if c in msub.columns]
        grouped = msub.groupby('iter')[available[1:]].mean().reset_index()

        print(f'\n{"="*100}')
        print(f'Method: {method}  |  lambda=0.5  |  alpha_fair={ALPHA}  |  averaged over seeds')
        print(f'{"="*100}')

        fmt_df = grouped.copy()
        for c in cos_cols:
            if c in fmt_df.columns:
                fmt_df[c] = fmt_df[c].map(lambda x: f'{x:+.4f}')
        for c in norm_cols:
            if c in fmt_df.columns:
                fmt_df[c] = fmt_df[c].map(lambda x: f'{x:.6f}')

        with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.width', 200):
            print(fmt_df[[c for c in display_cols if c in fmt_df.columns]].to_string(index=False))
else:
    print('No Experiment 2 iter data found.')
```

```python
# Cell 10: Start vs End comparison

if not iter_df_single.empty:
    df = iter_df_single.copy()
    compare_rows = []

    for method in sorted(df['config_name'].unique()):
        msub = df[df['config_name'] == method]
        iters = sorted(msub['iter'].unique())
        if len(iters) < 2:
            continue
        first_iter, last_iter = iters[0], iters[-1]

        for label, it in [('start', first_iter), ('end', last_iter)]:
            isub = msub[msub['iter'] == it]
            row = {'method': method, 'phase': label, 'iter': it}
            for c in cos_cols + norm_cols + ['loss_dec', 'loss_pred', 'loss_fair']:
                if c in isub.columns:
                    row[c] = isub[c].mean()
            compare_rows.append(row)

    compare_df = pd.DataFrame(compare_rows)
    print('=== Start vs End comparison (single lambda=0.5, averaged over seeds) ===')
    print()

    start = compare_df[compare_df['phase'] == 'start'].sort_values('method')
    end = compare_df[compare_df['phase'] == 'end'].sort_values('method')

    show_cols = ['method'] + [c for c in cos_cols + ['loss_dec', 'loss_pred', 'loss_fair'] if c in start.columns]

    print('--- Iteration 0 (all methods should be identical if same init) ---')
    print(start[show_cols].to_string(index=False, float_format='%.4f'))

    print(f'\n--- Final iteration ---')
    print(end[show_cols].to_string(index=False, float_format='%.4f'))
else:
    print('No Experiment 2 iter data found.')
```

## Unified Cross-Fairness Comparison

Compares all methods across fairness types using **Relative Fairness** =
fairness(λ=0.5) / fairness(λ=0), normalizing all metrics to a common scale.
Uses Exp 3 data only.

```python
# Cell 11: Unified cross-fairness comparison table

import warnings

all_stage_dfs = []
for ft in fairness_types:
    if ft not in results_by_fairness:
        continue
    sdf, _ = results_by_fairness[ft]
    df = sdf.copy()
    if 'fairness_type' not in df.columns:
        df['fairness_type'] = ft
    all_stage_dfs.append(df)

if not all_stage_dfs:
    print('No Exp 3 data available for unified comparison.')
else:
    combined = pd.concat(all_stage_dfs, ignore_index=True)
    has_norm = 'test_regret_normalized' in combined.columns
    regret_col = 'test_regret_normalized' if has_norm else 'test_regret'

    rows = []
    for (method, ft), grp in combined.groupby(['config_name', 'fairness_type']):
        lam0 = grp[grp['lambda'] == 0.0]
        lam05 = grp[grp['lambda'] == 0.5]
        if lam0.empty or lam05.empty:
            warnings.warn(f'Missing lambda=0 or lambda=0.5 for {method}/{ft}, skipping.')
            continue
        for seed in lam05['seed'].unique():
            s0 = lam0[lam0['seed'] == seed]
            s05 = lam05[lam05['seed'] == seed]
            if s0.empty or s05.empty:
                continue
            baseline = s0['test_fairness'].values[0]
            rel_fair = s05['test_fairness'].values[0] / baseline if baseline > 1e-12 else float('nan')
            rows.append({
                'method': method, 'fairness_type': ft, 'seed': seed,
                'norm_regret': s05[regret_col].values[0],
                'relative_fairness': rel_fair,
                'pred_mse': s05['test_pred_mse'].values[0],
            })

    if not rows:
        print('Could not compute relative fairness — missing lambda=0 baseline.')
        print('Re-run Exp 3 with lambdas=[0.0, 0.5] to generate baselines.')
    else:
        detail_df = pd.DataFrame(rows)
        agg = detail_df.groupby(['method', 'fairness_type']).agg(
            norm_regret_mean=('norm_regret', 'mean'), norm_regret_std=('norm_regret', 'std'),
            rel_fair_mean=('relative_fairness', 'mean'), rel_fair_std=('relative_fairness', 'std'),
            pred_mse_mean=('pred_mse', 'mean'), pred_mse_std=('pred_mse', 'std'),
            n_seeds=('seed', 'count'),
        ).reset_index()

        agg['Norm Regret'] = agg.apply(lambda r: f"{r['norm_regret_mean']:.4f} +/- {r['norm_regret_std']:.4f}", axis=1)
        agg['Rel. Fairness'] = agg.apply(
            lambda r: f"{r['rel_fair_mean']:.4f} +/- {r['rel_fair_std']:.4f}" if not np.isnan(r['rel_fair_mean']) else 'N/A', axis=1)
        agg['Pred MSE'] = agg.apply(lambda r: f"{r['pred_mse_mean']:.4f} +/- {r['pred_mse_std']:.4f}", axis=1)

        print('=' * 100)
        print(f'UNIFIED CROSS-FAIRNESS COMPARISON (alpha={ALPHA}, lambda=0.5)')
        print('Relative Fairness = fairness(lam=0.5) / fairness(lam=0)  [lower = more improvement]')
        print('=' * 100)
        display_cols = ['method', 'fairness_type', 'Norm Regret', 'Rel. Fairness', 'Pred MSE']
        print(agg[display_cols].sort_values(['fairness_type', 'norm_regret_mean']).to_string(index=False))

        # Rank consistency
        for metric_name, metric_col in [('norm_regret', 'norm_regret_mean'), ('relative_fairness', 'rel_fair_mean')]:
            print(f'\n{"="*80}')
            print(f'RANK CONSISTENCY (by {metric_name}, 1 = best)')
            print(f'{"="*80}')
            rd = {}
            for ft, ft_grp in agg.groupby('fairness_type'):
                ft_sorted = ft_grp.sort_values(metric_col).reset_index(drop=True)
                for rank, (_, r) in enumerate(ft_sorted.iterrows(), 1):
                    rd.setdefault(r['method'], {})[ft.upper()] = rank
            rt = pd.DataFrame.from_dict(rd, orient='index')
            rt.index.name = 'Method'
            rt['Avg Rank'] = rt.mean(axis=1)
            print(rt.sort_values('Avg Rank').to_string(float_format='%.1f'))
```

## Experiments 4-6: Hyperparameter Ablations

Load and display results from HP ablation experiments:
- **Exp 4**: Predictor architecture (linear / MLP / ResNet / FT-Transformer)
- **Exp 5**: Learning rate × steps grid
- **Exp 6**: Optimizer (SGD vs Adam)

Run these via: `python run_ablation.py --experiments 4 5 6`

```python
# Cell 12: Load HP ablation results

stage_hp_arch = load_csv('stage_hp_architecture.csv')
stage_hp_lr_steps = load_csv('stage_hp_lr_steps.csv')
stage_hp_optimizer = load_csv('stage_hp_optimizer.csv')
```

```python
# Cell 13: Exp 4 — Architecture comparison

if not stage_hp_arch.empty and 'hp_config' in stage_hp_arch.columns:
    print(f'=== Experiment 4: Predictor Architecture (alpha={ALPHA}) ===')
    print(f'Regret = normalized regret\n')

    for arch_name in sorted(stage_hp_arch['hp_config'].unique()):
        sub = stage_hp_arch[stage_hp_arch['hp_config'] == arch_name]
        tbl = make_summary_table(sub)
        tbl = tbl.sort_values('Regret_mean')
        tbl.insert(0, 'Rank', range(1, len(tbl) + 1))
        print(f'--- {arch_name} ---')
        display_cols = ['Rank', 'config_name', 'Regret', 'Fairness', 'Pred MSE']
        print(tbl[[c for c in display_cols if c in tbl.columns]].to_string(index=False))
        print()

    # Rank consistency
    print(f'{"="*80}')
    print('Ranking Consistency Across Architectures (by norm regret, 1 = best)')
    print(f'{"="*80}')
    rd = {}
    for arch_name in sorted(stage_hp_arch['hp_config'].unique()):
        sub = stage_hp_arch[stage_hp_arch['hp_config'] == arch_name]
        tbl = make_summary_table(sub).sort_values('Regret_mean').reset_index(drop=True)
        for rank, (_, r) in enumerate(tbl.iterrows(), 1):
            rd.setdefault(r['config_name'], {})[arch_name] = rank
    rt = pd.DataFrame.from_dict(rd, orient='index')
    rt.index.name = 'Method'
    rt['Avg Rank'] = rt.mean(axis=1)
    print(rt.sort_values('Avg Rank').to_string(float_format='%.1f'))
else:
    print('No Experiment 4 data found. Run: python run_ablation.py --experiments 4')
```

```python
# Cell 14: Exp 5 — LR x Steps

if not stage_hp_lr_steps.empty and 'hp_config' in stage_hp_lr_steps.columns:
    print(f'=== Experiment 5: LR x Steps (alpha={ALPHA}) ===')
    print(f'Regret = normalized regret\n')

    # Best HP per method
    print('--- Best HP config per method (lowest norm regret) ---')
    for method in sorted(stage_hp_lr_steps['config_name'].unique()):
        msub = stage_hp_lr_steps[stage_hp_lr_steps['config_name'] == method]
        has_norm = 'test_regret_normalized' in msub.columns
        rc = 'test_regret_normalized' if has_norm else 'test_regret'
        best_hp = msub.groupby('hp_config')[rc].mean().idxmin()
        best_val = msub.groupby('hp_config')[rc].mean().min()
        print(f'  {method:15s}  best={best_hp:30s}  norm_regret={best_val:.4f}')

    # Rank stability
    print(f'\n{"="*80}')
    print('Method ranking stability across LR x Steps configs')
    print(f'{"="*80}')
    method_ranks = {}
    for hp_label in sorted(stage_hp_lr_steps['hp_config'].unique()):
        sub = stage_hp_lr_steps[stage_hp_lr_steps['hp_config'] == hp_label]
        tbl = make_summary_table(sub).sort_values('Regret_mean').reset_index(drop=True)
        for rank, (_, r) in enumerate(tbl.iterrows(), 1):
            method_ranks.setdefault(r['config_name'], []).append(rank)
    stab = [{'method': m, 'mean_rank': np.mean(r), 'std_rank': np.std(r),
             'min_rank': min(r), 'max_rank': max(r)}
            for m, r in sorted(method_ranks.items())]
    print(pd.DataFrame(stab).sort_values('mean_rank').to_string(index=False, float_format='%.2f'))
else:
    print('No Experiment 5 data found. Run: python run_ablation.py --experiments 5')
```

```python
# Cell 15: Exp 6 — Optimizer comparison

if not stage_hp_optimizer.empty and 'hp_config' in stage_hp_optimizer.columns:
    print(f'=== Experiment 6: Optimizer Comparison (alpha={ALPHA}) ===')
    print(f'Regret = normalized regret\n')

    for opt in sorted(stage_hp_optimizer['hp_config'].unique()):
        sub = stage_hp_optimizer[stage_hp_optimizer['hp_config'] == opt]
        tbl = make_summary_table(sub)
        tbl = tbl.sort_values('Regret_mean')
        tbl.insert(0, 'Rank', range(1, len(tbl) + 1))
        print(f'--- {opt.upper()} ---')
        display_cols = ['Rank', 'config_name', 'Regret', 'Fairness', 'Pred MSE']
        print(tbl[[c for c in display_cols if c in tbl.columns]].to_string(index=False))
        print()

    # Side-by-side
    opts = sorted(stage_hp_optimizer['hp_config'].unique())
    if len(opts) == 2:
        print(f'{"="*80}')
        print(f'{opts[0].upper()} vs {opts[1].upper()} — per method delta')
        print(f'{"="*80}')
        tbl0 = make_summary_table(stage_hp_optimizer[stage_hp_optimizer['hp_config'] == opts[0]]).set_index('config_name')
        tbl1 = make_summary_table(stage_hp_optimizer[stage_hp_optimizer['hp_config'] == opts[1]]).set_index('config_name')
        for m in sorted(set(tbl0.index) & set(tbl1.index)):
            r0, r1 = tbl0.loc[m, 'Regret_mean'], tbl1.loc[m, 'Regret_mean']
            delta = r1 - r0
            winner = opts[1].upper() if delta < 0 else opts[0].upper()
            print(f'  {m:15s}  {opts[0]}={r0:.4f}  {opts[1]}={r1:.4f}  delta={delta:+.4f}  winner={winner}')
else:
    print('No Experiment 6 data found. Run: python run_ablation.py --experiments 6')
```
